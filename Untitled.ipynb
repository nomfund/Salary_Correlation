# a bar graph to show the different average salaries of men and women.

```python
_, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 10))

sns.countplot(x='salary', data=salary_data, ax=axes[0]);
sns.countplot(x='male', data=salary_data, ax=axes[1]);
```

# Now we are looking for features that are strongly correlated with salary and store them in golden_features_list 

```python
df_num_corr= df_num.corr()['salary'][:1]
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Salary:\n{}".format(len(golden_features_list), golden_features_list))
```
